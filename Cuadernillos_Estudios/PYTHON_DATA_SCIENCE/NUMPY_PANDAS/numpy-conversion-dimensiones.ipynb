```python
## DEFINIR EL TIPO DE DATO A LA HORA DE CREAR EL ARRAY

arr=np.array([1,2,3,4],dtype="float64") ## DEFINIMOS QUE LO COMPONEN FLOTANTES DE 64 BITS
arr.dtype
```

```python
### CAMBIAR EL TIPO DE DATOS:

arr=arr.astype(np.int64) ### LO DEFINIMOS DIRECTAMENTE DESDE LA LIBRERIA
arr.dtype
```

```python
### LOGICA BOOLEANA 
### LOS CEROS SIEMPRE ESTAN EN FALSO Y LOS DEMAS NUMEROS SON VERDADERO:

arr_2=np.array([0,1,2,3,4])
arr_2=arr_2.astype(np.bool_) ## LOS DEFINIMOS COMO BOOLEANOS

arr_2
```

## DIMENSIONES

```python
### DEFINIMOS UN ESCALAR
scalar=np.array(42)
print(scalar.ndim) ## COMANDO PARA VER LAS DIMENSIONES DEL ARRAY
scalar.shape
```

```python
## OBSERVEMOS CON UNA LISTA
vector=np.array([1,2,3])
vector.ndim ## UNA DIMENSION
```

```python
matriz=np.array([[1,2,3],[1,2,4]])
print(matriz)
print(matriz.ndim)
```

```python
### TENSORES

tensor=np.array([[[1,2,3],[1,2,4]]])
print(tensor.ndim)
tensor.shape  ##(un ejemplar de 2,3 como matriz)

```

```python
### DEFINIR DIMENSIONES
VECTOR=np.array([1,2,3],ndmin=10) ### ME DEFINE DIRECTAMENTE LAS DIMENSIONES QUE DEBE TENER
print(VECTOR)
VECTOR.ndim
```

OJO EL CERO A NIVEL DE NUMPY Y EN GENERAL PARA PYTHON REPRESENTA FILAS, Y EL UNO REPRESENTA COLUMNAS, POR LO TANTO EL AXIS=0, ES PARA DEFINIR LOS ELEMENTOS EN FILAS, Y EL AXIS=1 ES PARA DEFINIRLOS A NIVEL DE COLUMNAS LOS ELEMENTOS.

```python
expand=np.expand_dims(np.array([1,2,3]),axis=0)
## AL ARREGLO LE VOY A EXPANDIR UNA DIMENSION MAS
print(expand) ## LE AGREGO UNA DIMENSION MAS

expand.ndim
```
