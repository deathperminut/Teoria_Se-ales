```python
import numpy as np
```

```python
arr=np.linspace(1,10,10,dtype="int8")
arr
```

```python
indices_cond= arr > 5 ## ME RETORNA UNA LISTA BOOLEANA ME AHORRA MUCHO TRABAJO
indices_cond
```

```python
### AHORA SI QUIERO VER LOS VALORES COMO TAL PUEDO HACER LO SIGUIENTE:

arr[indices_cond] ## ENTREGANDOME SOLO LOS QUE LA CUMPLEN
```

```python
### ENTONCES PUEDO JUGAR CON ESOS CONDICIONES DE FORMA LOGICA
arr[(arr>5) & (arr<9)] ## EL & ES UN AND LOGICO 
```

```python
arr[arr>5]=99
arr

## ASI FILTRAMOS ESTRUCTURAS DE DATOS
```
