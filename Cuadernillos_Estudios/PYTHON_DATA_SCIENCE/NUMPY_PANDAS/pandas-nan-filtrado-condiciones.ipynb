```python
import numpy as np
import pandas as pd
```

```python
### CREAREMOS UN DICCIONARIO
Dict={"Col1":[1,2,3,np.nan],
 "Col2":[4,np.nan,6,7],
 "Col3":["a","b","c",None]}
Dict
```

```python
### AHORA CREAMOS UN DATAFRAME DE DICHO DICCIONARIO

df=pd.DataFrame(Dict)
df
```

```python
### MANEJO DE VALORES NULOS
### None para valores de tipo texto y NaN valores de tipo numerico

df.isnull()
```

```python
# PUEDO CONVERTIR ESO A UNA MATRIZ DE UNOS Y CEROS DE LA SIGUIENTE FORMA
df.isnull()*1
```

```python
## UNA VEZ IDENTIFICADOS LOS VALORES NULOS, PODEMOS LLENAR LOS VALORES NULOS DE LA SIGUIENTE MANERA
df.fillna("Missing")
### DE ESTE FORMA PUEDO FILTRAR 
```

```python
df.interpolate() ### ASUME UN VALOR PARA LOS NULOS, MEDIANTE UNA SERIE
### PARA LOS TEXTOS NO LOGRA IDENTIFICAR.
```

```python
df.dropna() ## ELIMINA LOS VALORES NULOS ## ELIMINA DIRECTAMENTE LOS SUJETOS POR COMPLETO
```
