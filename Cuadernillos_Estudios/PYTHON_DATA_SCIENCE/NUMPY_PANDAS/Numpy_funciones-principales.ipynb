### CONCATENAR

```python
### CONCATENAR 
a=np.array([[1,2],[3,4]]) ## CREAMOS UNA MATRIZ
a
```

```python
b=np.array([5,6])
b
```

```python
print(a.ndim)
print(b.ndim)
```

```python
##PARA CONCATENAR DEBEN TENER LA MISMA DIMENSION
b=np.expand_dims(b,axis=0)
print(b.ndim)
```

```python
## AHORA SI CONCATENAMOS
np.concatenate((a,b),axis=0)
```

```python
#TRANSPUESTA
b.T
## SI QUIERO AGREGARLA A MANERA DE COLUMNA
np.concatenate((a,b.T),axis=1)
```
