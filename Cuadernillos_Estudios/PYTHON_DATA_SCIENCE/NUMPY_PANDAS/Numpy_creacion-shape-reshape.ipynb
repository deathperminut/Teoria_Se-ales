### SHAPE Y RESHAPE

```python
arr=np.random.randint(1,10,(3,2))

arr.shape ## ME INDICA CUANTAS FILAS Y CUANTAS COLUMNAS TENGO
```

```python
arr.reshape(1,6) ## ME CAMBIA LA FORMA A UNA FILA Y 6 COLUMNAS


```

```python
### NO TENGO PORQUE APLICARLO DIRECTAMENTE A NUESTRO ARRAY SINO QUE PUEDO AGREGARLA 
### COMO UN PARAMETRO
np.reshape(arr,(1,6))
```

```python
### PUEDO DEFINIR SI PRIORIZAR EL ORDEN DE LAS FILAS O COLUMNAS
np.reshape(arr,(2,3),"C") ### C DE LENGUAJE C ES DECIR SIGUE LOS REGIMENES DE C
```

```python
np.reshape(arr,(2,3),"F")     ##EN FORTRANT  PRIORIZA COLUMNAS
```

```python
np.reshape(arr,(2,3),"A") ## EN MI SISTEMA EN MI LAPTOP ## ESCOGE UNO DE LOS 2 ES DECIR
### ESTA EN MODO C

### YO NO PUEDO CAMBIAR HACIA UNA FORMA QUE NO SE PUEDA FORMAR CON LA DISTRUCIÓN DE LOS ELEMENTOS
### ORIGINAL.

### NO PUEDO INVERTARME LOS DATOS
```
