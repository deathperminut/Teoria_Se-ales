```python
### IMPORTAMOS LA DATA MMEDIANTE ESTE URL:
### https://www.kaggle.com/alarcon7a/cnn-data-sources
### CUADERNILLO EN KAGGLE 


import numpy as np
import matplotlib.pyplot as plt


############# TRATRAMIENTO DE IMAGENES##############################################################
import scipy.ndimage as nd  ## LIBRERIA QUE SIRVE PARA TRABAJAR CONVOLUCIONES CON IMAGENES 
from skimage import io,color
####################################################################################################
```

```python
## IMPORTAMOS LA IMAGEN YO MEDIANTE SU URL ASOCIADA:
img=io.imread("../input/cnn-data-sources/yo.png")
```

```python
img
```

```python
plt.imshow(img) ## TENEMOS UNA FOTO 
img.shape ## TIENE 3 CAPAS
```

```python
### AHORA LO MEJOR SIEMPRE ES TRABAJAR EN ESCALA DE GRISES, MIENTRAS NO SEA IMPORTANTE EL COLOR

img_gray=color.rgb2gray(img) ### COMANDO PARA CONVERTIR LA IMAGEN A ESCALA DE GRISES SI OBSERVAMOS SUS DIMENSIONES
## ES DE UNA SOLA CAPA
plt.imshow(img_gray)
img_gray.shape ### SOLO TIENE UNA DIMENSION ESCALA DE GRISES.
```

```python
### CREAMOS NUESTRO KERNEL
### CREAMOS NUESTRO KERNEL PARA DETERMINAR LINEAS VERTICALES.


kernel_vertical=np.array([[-1,0,1],
                          [-1,0,1],
                          [-1,0,1]])
```

```python
## LA LIBRERIA ND ME PERMITE REALIZAR CONVOLUCIONES EN IMAGENES

img_kernel=nd.convolve(img_gray,kernel_vertical)
```

```python
### GRAFICAMOS ENTONCES
fig,axes= plt.subplots(1,2,figsize=(15,10))
axes[0].imshow(img_gray,cmap=plt.cm.gray)
axes[0].set_title("original")
axes[0].axis("off")

axes[1].imshow(img_kernel,cmap=plt.cm.gray)
axes[1].set_title("Kernel")
axes[1].axis("off")
```

PODEMOS OBSERVAR QUE EL KERNEL PERMITIO DETECTAR LINEAS VERTICALES DENTRO DE LA MISMA IMAGEN, CON ESTE PRINCIPIO SE ANALIZA INFORMACIÃ“N ESPECIFICA DENTRO DE LA IMAGEN.

```python
### APLIQUEMOS OTRO FILTRO DISTINTO (KERNEL)

kernel2=np.array([[0,0,0],
                  [0,1,0],
                  [0,0,0]])

### ME MUESTRA EXACTAMENTE LA MISMA IMAGEN.

### INTENTO DE BORDES HORIZONTALES
kernel3=np.array([[-1,-1,-1],
                  [0,0,0],
                  [1,1,1]])
```

```python
img_kernel=nd.convolve(img_gray,kernel3)
### GRAFICAMOS ENTONCES
fig,axes= plt.subplots(1,2,figsize=(15,10))
axes[0].imshow(img_gray,cmap=plt.cm.gray)
axes[0].set_title("original")
axes[0].axis("off")

axes[1].imshow(img_kernel,cmap=plt.cm.gray)
axes[1].set_title("Kernel")
axes[1].axis("off")
```
