```python
### IMPORTAMOS LA DATA MMEDIANTE ESTE URL:
### https://www.kaggle.com/alarcon7a/cnn-data-sources
### CUADERNILLO EN KAGGLE 

import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im=io.imread("../input/cnn-data-sources/Mara.jpeg") ## GUARDAMOS LA IMAGEN
im ## UN CONJUNTO DE VALORES MATRICIALES
```

```python
im.shape ## UNA IMAGEN CON UNA FORMA DE 800,600 Y 3 CAPAS ES DECIR ESTA A COLOR.
```

```python
plt.imshow(im) 
```

```python
### TENSORFLOW PUEDE RECIBIR TENSORES DE DATOS, POR ESO ES QUE PUEDE MANEJAR IMAGENES
### OBSERVEMOS AHORA CADA UNA DE LAS CAPAS:
r=im[:,:,0]
g=im[:,:,1]
b=im[:,:,2]
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,15))
ax[0].imshow(r) ## AL SEPARAR LAS CAPAS YA TRABAJO ES EN ESCALA DE GRISES PARA QUE SE VEA MEJOR USO EL 
## COMANDO CMAP="GRAY"
ax[1].imshow(g)
ax[2].imshow(b)
plt.show()

plt.imshow(r,cmap="gray")
```

```python
## AHORA YA TENGO MATRICES DE NUMPY, ENTONCES YO PUEDO SIMPLEMENTE
## DECIRLE QUE ME MUESTRA LA TRANSPUESTA POR EJEMPLO

plt.imshow(r.T,cmap="gray")
```

```python
### VAMOS A OBSERVAR QUE DIMENSIÓN TIENE R
r.shape
```

```python
## CREAREMOS UN ARREGLO DE DICHO TAMAÑO:

aux_dim=np.zeros([800,600])

Imag_R=np.dstack((r,aux_dim,aux_dim)).astype(np.uint8) ## FORMO MI MATRIZ DE 3 CAPAS PERO ENTONCES SOLO HABILITE 
## MIS VALORES DE ROJO
Imag_g=np.dstack((aux_dim,g,aux_dim)).astype(np.uint8) ## FORMO MI MATRIZ DE 3 CAPAS
Imag_b=np.dstack((aux_dim,aux_dim,b)).astype(np.uint8) ## FORMO MI MATRIZ DE 3 CAPAS

## GRAFICAMOS
fig,ax=plt.subplots(1,3,figsize=(15,15))
ax[0].imshow(Imag_R) ## AL SEPARAR LAS CAPAS YA TRABAJO ES EN ESCALA DE GRISES PARA QUE SE VEA MEJOR USO EL 
## COMANDO CMAP="GRAY"
ax[1].imshow(Imag_g)
ax[2].imshow(Imag_b)
plt.show()
```

```python
## AHORA SI INVIERTO LA POLARIDAD 
im_neg=255-im
##BAJARLE LA RESOLUCIÓN
im32=(im//32)*32
im128=(im//128)*128


## GRAFICAMOS
fig,ax=plt.subplots(1,3,figsize=(15,15))
ax[0].imshow(im_neg) ## AL SEPARAR LAS CAPAS YA TRABAJO ES EN ESCALA DE GRISES PARA QUE SE VEA MEJOR USO EL 
## COMANDO CMAP="GRAY"
ax[1].imshow(im32)
ax[2].imshow(im128)
plt.show()
```
